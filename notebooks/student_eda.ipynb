```python
import pandas as pd
df = pd.read_csv("../data/StudentPerformanceFactors.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df_clean = df.copy()
```

```python
missing = df_clean.isnull().sum()
missing
```

```python
df_clean = df_clean.dropna()
```

```python
df_clean.isnull().sum()
```

```python
df_clean.describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (7,5))
plt.scatter(df_clean["Attendance"],df_clean["Exam_Score"])
plt.xlabel("Attendance (%)")
plt.ylabel("Exam Score")
plt.title("Attendance vs Exam Score")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df_clean["Hours_Studied"], df_clean["Exam_Score"])
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Hours Studied vs Exam Score")
plt.show()
```

```python
df_clean.groupby("Teacher_Quality")["Exam_Score"].mean()
```

```python
df_clean.groupby("Teacher_Quality")["Exam_Score"].mean().plot(kind="bar")
plt.title("Average Exam Score by Teacher Quality")
plt.ylabel("Average Exam Score")
plt.show()
```
